# Goal

Convert the file `subject_code_lookup.json` to a more usable Excel file.

```python
import pandas as pd
```

```python
codes = pd.read_json('subject_code_lookup.json')
codes.info()
codes.head()
```

```python
df = pd.DataFrame(
    {
        'Code': list(codes.columns),
        'Area': codes.loc['area', :],
        'Abbreviation': codes.loc['abbrev', :]        
    }
).reset_index(drop=True)
```

```python
df
```

```python
df.to_excel('/media/Bibliometrics/matt analysis/subject_codes_lookup.xlsx', index=False)
```
